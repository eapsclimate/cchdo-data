```python
from pymongo import MongoClient
from tqdm import tqdm
import folium
```

```python
rcs_hyd = MongoClient().eddies.rcs_hyd
```

```python
rcs_hyd.find_one()
```

```python
map_test = folium.Map(location=[0, 0], zoom_start=1)
```

```python
for document in tqdm(rcs_hyd.find()):
    eddy_lat = document['eddy_loc']['coordinates'][1]
    eddy_lon = document['eddy_loc']['coordinates'][0]
    eddy_coo = [eddy_lat, eddy_lon]
    eddy_rad = document['eddy_rad']
    expo_lat = document['expo_loc'][1]
    expo_lon = document['expo_loc'][0]
    expo_coo = [expo_lat, expo_lon]
    eddy_inf = "eddy-id: " + document['EDDYCODE']
    expo_inf = "expo-id: " + document['EXPOCODE'] + '-' + document['expo_sta'] + '-' + document['expo_cas']
    folium.folium.Marker(eddy_coo, popup=eddy_inf).add_to(map_test)
    folium.CircleMarker(eddy_coo, radius=eddy_rad).add_to(map_test)
    folium.folium.Marker(expo_coo, popup=expo_inf).add_to(map_test)
```
