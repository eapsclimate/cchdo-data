```python
from ctd2xray import cchdo
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import xarray as xr
```

```python
%matplotlib inline
```

```python
ds_60 = xr.open_dataset('/home/cz2397/data/cchdo-ctd/351_09FA20010524_nc_ctd/09FA20010524_00060_00001_ctd.nc')
```

```python
path = glob('/home/cz2397/data/cchdo-ctd/351_09FA20010524_nc_ctd/09FA20010524_000[5-6][0-9]_*_ctd.nc')
```

```python
pressure = np.arange(1., 1000., 1.)
```

```python
ds = cchdo.open_cchdo_as_mfdataset(path, pressure)
```

```python
ds = ds.swap_dims({'time': 'latitude'})
```

```python
lat = ds.latitude.values
lon = ds.longitude.values
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
plt.plot(lon, lat, '.')
plt.xlabel('$longitude$')
plt.ylabel('$latitude$')
```

```python
lat_min = ds.latitude.values.min()
lat_min
```

```python
lat_max = ds.latitude.values.max()
lat_max
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ds.temperature_i.T.plot(yincrease=False, ax=ax, vmin=2, vmax=20, cmap='bwr')
plt.axvline(ds_60.latitude.values[0], color='w')
plt.xlabel('$latitude$')
plt.ylabel('$pressure$')
plt.xlim([lat_min, lat_max])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ds.salinity_i.T.plot(yincrease=False, ax=ax, vmin=34, vmax=36, cmap='bwr')
plt.axvline(ds_60.latitude.values[0], color='w')
plt.xlabel('$latitude$')
plt.ylabel('$pressure$')
plt.xlim([lat_min, lat_max])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ds.oxygen_i.T.plot(yincrease=False, ax=ax, vmin=0, vmax=300, cmap='bwr')
plt.axvline(ds_60.latitude.values[0], color='w')
plt.xlabel('$latitude$')
plt.ylabel('$pressure$')
plt.xlim([lat_min, lat_max])
```
