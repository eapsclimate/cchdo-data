```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import xarray as xr
%matplotlib inline
```

```python
path = '/home/cz2397/repositories/cchdo-data/69001/69001_prof.nc'
ds = xr.open_dataset(path)
```

```python
time = ds.JULD.values[0]
lon = ds.LONGITUDE.values[0]
lat = ds.LATITUDE.values[0]
p = ds.PRES.values[0]
t = ds.TEMP.values[0]
```

```python
sns.set()
plt.rcParams['figure.figsize'] = (12, 12)
my_map = Basemap(projection='ortho', resolution='c',
                 lon_0=lon.mean(), lat_0=lat.mean())
my_map.drawmapboundary('none')
my_map.drawmeridians(np.arange(0, 360, 10), color='#ffffff')
my_map.drawparallels(np.arange(-90, 90, 10), color='#ffffff')
my_map.fillcontinents()
lon_argo, lat_argo = my_map(lon, lat)
my_map.plot(lon_argo, lat_argo, '.', color='#0080ff')
plt.show()
```

```python
sns.set()
plt.rcParams['figure.figsize'] = (12, 12)
plt.plot(t, p)
plt.xlabel('temperature')
plt.ylabel('pressure')
plt.ylim([1000, 0])
plt.show()
```
