### packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import xarray as xr
```

```python
%matplotlib inline
```

### data

```python
DSM3 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00005_ctd.nc')
DSM2 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00006_ctd.nc')
DSM1 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00007_ctd.nc')
DS00 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00008_ctd.nc')
DSP1 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00009_ctd.nc')
DSP2 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00010_ctd.nc')
DSP3 = xr.open_dataset\
('/home/cz2397/data/cchdo-ctd/12_33KI121_1_nc_ctd/prs2_00002_00011_ctd.nc')
```

```python
DFM3 = DSM3.to_dataframe()
DFM2 = DSM2.to_dataframe()
DFM1 = DSM1.to_dataframe()
DF00 = DS00.to_dataframe()
DFP1 = DSP1.to_dataframe()
DFP2 = DSP2.to_dataframe()
DFP3 = DSP3.to_dataframe()
```

### pressure

```python
pressure_M3 = DSM3.indexes['pressure'].values
pressure_M2 = DSM2.indexes['pressure'].values
pressure_M1 = DSM1.indexes['pressure'].values
pressure_00 = DS00.indexes['pressure'].values
pressure_P1 = DSP1.indexes['pressure'].values
pressure_P2 = DSP2.indexes['pressure'].values
pressure_P3 = DSP3.indexes['pressure'].values
```

### temperature

```python
temperature_M3 = DFM3.temperature
temperature_M2 = DFM2.temperature
temperature_M1 = DFM1.temperature
temperature_00 = DF00.temperature
temperature_P1 = DFP1.temperature
temperature_P2 = DFP2.temperature
temperature_P3 = DFP3.temperature
```

```python
plt.rcParams['figure.figsize'] = (20, 8)
f, ax = plt.subplots(1, 7, sharex=True, sharey=True)
ax[0].plot(temperature_M3, pressure_M3)
ax[1].plot(temperature_M2, pressure_M2)
ax[2].plot(temperature_M1, pressure_M1)
ax[3].plot(temperature_00, pressure_00, 'c')
ax[4].plot(temperature_P1, pressure_P1)
ax[5].plot(temperature_P2, pressure_P2)
ax[6].plot(temperature_P3, pressure_P3)
plt.xlim([0, 30])
plt.ylim([0, 1000])
plt.gca().invert_yaxis()
plt.xlabel('$temperature$')
plt.ylabel('$pressure$')
```

### salinity

```python
salinity_M3 = DFM3.salinity
salinity_M2 = DFM2.salinity
salinity_M1 = DFM1.salinity
salinity_00 = DF00.salinity
salinity_P1 = DFP1.salinity
salinity_P2 = DFP2.salinity
salinity_P3 = DFP3.salinity
```

```python
plt.rcParams['figure.figsize'] = (20, 8)
f, ax = plt.subplots(1, 7, sharex=True, sharey=True)
ax[0].plot(salinity_M3, pressure_M3)
ax[1].plot(salinity_M2, pressure_M2)
ax[2].plot(salinity_M1, pressure_M1)
ax[3].plot(salinity_00, pressure_00, 'c')
ax[4].plot(salinity_P1, pressure_P1)
ax[5].plot(salinity_P2, pressure_P2)
ax[6].plot(salinity_P3, pressure_P3)
plt.xlim([34, 36])
plt.ylim([0, 1000])
plt.gca().invert_yaxis()
plt.xlabel('$salinity$')
plt.ylabel('$pressure$')
```
